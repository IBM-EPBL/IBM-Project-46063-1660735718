```python
for index in range(1,5):
    img = Image.open(r"D:\Images\image_" +str(index)+'.jpg').convert("L")
    img = img.resize((28,28))
    img2arr = np.array(img)
    img2arr = img2arr.reshape(1,28,28,1)
    y_pred = model.predict(img2arr)
    print(y_pred)
    print(f"The image_{index} predicted as value {np.argmax(y_pred)}")
```
